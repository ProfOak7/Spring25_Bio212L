```python
import pandas as pd
import statistics
from statistics import mean, stdev
from scipy.stats import sem
import scipy.stats as stats
import matplotlib.pyplot as plt
#include if using a jupyter notebook, remove if using a .py file
%matplotlib inline  
```

```python
#Optional - Set figure resolution to 300dpi
import matplotlib as mpl
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
#Import the excel file
df = pd.read_excel (r'./Lab2.xlsx', index_col = 0) 
df
```

```python
df = df.transpose()
df
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'Forks', marker = 'o')
df.plot.line(ax = ax, y = 'Spoons', marker = 'o')
df.plot.line(ax = ax, y = 'Straws', marker = 'o')
df.plot.line(ax = ax, y = 'Forceps', marker = 'o')
df.plot.line(ax = ax, y = 'Tongue Depressors', marker = 'o')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.set_xlabel("Generation Number")
ax.set_ylabel("Population Size")
ax.set_title("Population Size Over Time")
```
